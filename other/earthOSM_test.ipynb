```python
from earth_osm import eo
import geopandas as gpd
import pandas as pd
```

```python
# Define the region 
region_list = ['brandenburg']

# Define the features to extract
feature_list = [
    'substation'
    #,'line', 'generator'
]

# Save the data
eo.save_osm_data(
    region_list=region_list,
    primary_name='power',
    feature_list=feature_list,
    update=False,  # Set to True if you want to update from OSM
    mp=True,  # Use multiprocessing for faster processing
    data_dir='osm_data_test',  # Directory to store downloaded data
    out_format=['geojson'],  # Output formats
    out_aggregate=False  # Set to True if you want aggregated results
)

print("Data download complete!")
```
